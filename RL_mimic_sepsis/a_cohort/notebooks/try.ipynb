```python
bins = [-np.inf, 0, 125, 250, 500, np.inf]
labels = [
    "Level 0: =0",
    "Level 1: (0, 125 )",
    "Level 2: [125 , 250 )",
    "Level 3: [250 , 500 )",
    "Level 4: ≥500 "
]

df['level'] = pd.cut(df['o:input_4hourly'], bins=bins, labels=labels, right=False)
df['level'].value_counts().sort_index()
```
