```python
import os
import pandas as pd
```

```python
# Read cohort files
hours = [1, 2, 4, 8]
full_data_files = {}
full_zs_dict = {}
for i in hours:
    save_dir = f'../data/newaction_{i}h_old'
    full_data_files[i] = os.path.join(save_dir, 'sepsis_final_data_withTimes_newActions.csv')
    full_zs_dict[i] = pd.read_csv(full_data_files[i])
    print(f'Successfully read {full_data_files[i]}')
    
    # if i == 1 or i == 2:
    #     full_data_files[i] = os.path.join(save_dir, 'sepsis_final_data_withTimes_filled.csv')
    #     full_zs_dict[i] = pd.read_csv(full_data_files[i])
    #     print(f'Successfully read {full_data_files[i]}')
    # else:
    #     full_data_files[i] = os.path.join(save_dir, 'sepsis_final_data_withTimes.csv')
    #     full_zs_dict[i] = pd.read_csv(full_data_files[i])
    #     print(f'Successfully read {full_data_files[i]}')

```

```python
for count, full_zs in full_zs_dict.items(): # key = count (hour), value = full_zs
    num_unique_traj = full_zs['traj'].nunique()
    print(f"Number of Trajectory of {count} hour's cohort: {num_unique_traj}")

    num_rows = full_zs.shape[0]
    print(f"Number of Steps of {count} hour's cohort: {num_rows}")

    lengths = full_zs.groupby('traj')['step'].count()  
    mean_len   = lengths.mean()
    median_len = lengths.median()
    q25, q75   = lengths.quantile([0.25, 0.75])
    iqr_len    = q75 - q25                             # inter-quartile range

    print(f"Mean   steps/trajectory: {mean_len:.2f}")
    print(f"Median steps/trajectory: {median_len:.2f}")
    print(f"IQR    steps/trajectory: {iqr_len:.2f}  (Q1={q25:.2f}, Q3={q75:.2f})")
    print()
```
