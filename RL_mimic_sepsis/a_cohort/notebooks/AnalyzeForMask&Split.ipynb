```python
"""This notebook explores the extent of set overlapping on icustayid across timesteps. Including mask. 
"""

import sys

import pandas as pd
import numpy as np

sys.path.append('F:/time_step/OfflineRL_FactoredActions')
from RL_mimic_sepsis.utils.timestep_util import timestep_list
```

```python
dfs = {}

for timestep in timestep_list:
    filename = (rf'F:\time_step\OfflineRL_FactoredActions\RL_mimic_sepsis\data'
                rf'\data_asNormThreshold_dt{timestep}h'
                rf'\sepsis_final_data_FILLED_withTimes_dt{timestep}h_asNormThreshold.csv')
    dfs[f'df_{timestep}h'] = pd.read_csv(filename)
```

```python
for timestep in timestep_list:
    print(f'Number of trajectories for dt={timestep}h before masking lengths <= 1:',
        len(dfs[f'df_{timestep}h']['m:icustayid'].unique()))
```

```python
# Remove trajectories that lengths <= 1.
for timestep in timestep_list:
    for traj in dfs[f'df_{timestep}h']['traj'].unique():
        if dfs[f'df_{timestep}h']['traj'].eq(traj).sum() <= 1:
            dfs[f'df_{timestep}h'] = dfs[f'df_{timestep}h'][~dfs[f'df_{timestep}h']['traj'].eq(traj)]
```

```python
for timestep in timestep_list:
    print(f'Number of trajectories for dt={timestep}h after  masking lengths <= 1:',
        len(dfs[f'df_{timestep}h']['m:icustayid'].unique()))
```

```python
id_sets = {
    t: set(dfs[f"df_{t}h"]["m:icustayid"].unique())
    for t in timestep_list
}
```

```python
# Passing each set as a separate argument to 'set.intersection'.
# Find intersections of 4 sets.
common_ids = set.intersection(*id_sets.values())
print(f"{len(common_ids)} icustays appear in every table")
```

```python
shared_ids_path = rf"F:\time_step\OfflineRL_FactoredActions\RL_mimic_sepsis\data\shared_icustayids_plus_mask.csv"
pd.Series(sorted(common_ids), name="m:icustayid").to_csv(shared_ids_path, index=False)
print('Saved sucessfully.')
```
