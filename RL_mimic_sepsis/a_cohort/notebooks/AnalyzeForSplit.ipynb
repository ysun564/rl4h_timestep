```python
"""This notebook explores the extent of set overlapping on icustayid across timesteps. 
"""

import sys

import pandas as pd
import numpy as np

sys.path.append('F:/time_step/OfflineRL_FactoredActions')
from RL_mimic_sepsis.utils.timestep_util import timestep_list
```

```python
dfs = {}

for timestep in timestep_list:
    filename = (rf'F:\time_step\OfflineRL_FactoredActions\RL_mimic_sepsis\data'
                rf'\data_asNormThreshold_dt{timestep}h'
                rf'\sepsis_final_data_FILLED_withTimes_dt{timestep}h_asNormThreshold.csv')
    dfs[f'df_{timestep}h'] = pd.read_csv(filename)
```

```python
id_sets = {
    t: set(dfs[f"df_{t}h"]["m:icustayid"].unique())
    for t in timestep_list
}
```

```python
for t, s in id_sets.items():
    print(f"{t}-hour table has {len(s)} unique ICU stays")
```

```python
# Test if dt=8 is a subset of dt=1. The answer is NO.
is_subset = id_sets[8].issubset(id_sets[1])
print("8-hour ⊆ 1-hour ?", is_subset)
```

```python
# Passing each set as a separate argument to 'set.intersection'.
common_ids = set.intersection(*id_sets.values())
print(f"{len(common_ids)} icustays appear in every table")
```

```python
shared_ids_path = rf"F:\time_step\OfflineRL_FactoredActions\RL_mimic_sepsis\data\shared_icustayids.csv"
pd.Series(sorted(common_ids), name="m:icustayid").to_csv(shared_ids_path, index=False)
print('Saved sucessfully.')
```
