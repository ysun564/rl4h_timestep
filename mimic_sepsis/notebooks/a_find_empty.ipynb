```python
# Determine if there're empty values in these 2 columns.
COLS = ['o:Creatinine', 'o:PaO2_FiO2']

def column_is_empty(col: pd.Series) -> bool:
    """If all values of the column are NaN, empty strings
    or whitespace, return True, else return False.
    """

    # If all of the columns are empty, return True.
    if col.isna().all():
        return True
    
    # checks for empty strings or whitespace.
    return (col.astype(str).str.strip() == '').all()

empty_flags = [column_is_empty(df[c]) for c in COLS]

empty_flags
```
